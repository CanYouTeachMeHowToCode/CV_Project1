```python
# Define the transforms to be applied to the images
transform = T.Compose([
    T.Resize(256), 
    T.CenterCrop(224),
    T.ToTensor(),
    T.Normalize(mean=(0.485, 0.456, 0.406), 
                std=(0.229, 0.224, 0.225)), # normalize the pixel values according to the mean and std of ImageNet
])

# Load the dataset and apply the defined transforms with class balancing
numHuman, numAI = len(os.listdir("../data_sourcing/images/selected_images/human")), len(os.listdir("../data_sourcing/images/selected_images/generated_images"))
print("number of images from human: {numHuman} \n number of images from AI: {numAI}".format(numHuman=numHuman, numAI=numAI))

dataset = torchvision.datasets.ImageFolder(root='../data_sourcing/images/selected_images', transform=transform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=8, shuffle=True, num_workers=4)
for (images, labels) in dataloader:
    print(labels)
# TODO: class balancing 
```
